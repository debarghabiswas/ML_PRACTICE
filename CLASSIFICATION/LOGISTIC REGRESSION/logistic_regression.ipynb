IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORTING THE DATASET

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
x
```

```python
y
```

SPLITTING THE DATSET INTO THE TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
xtrain
```

```python
xtest
```

TRAINING THE LOGISTIC REGRESSION MODEL ON THE TRAINING SET

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(xtrain, ytrain)
```

PREDICTING A NEW RESULT

```python
classifier.predict(sc.transform([[30, 87000]]))
```

PREDICTING THE TEST SET RESULTS

```python
ypred = classifier.predict(xtest)
print(np.concatenate((ypred.reshape(len(ypred),1), ytest.reshape(len(ytest),1)), axis=1))
```

MAKING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(ytest, ypred)
print(cm)

'''
HERE, 
52 IS THE CORRECT NO. OF PREDICTION OF CLASS ZERO(CUSTOMERS WHO DIDN'T BOUGHT THE SUV)
6 IS THE INCORRECT PREDICTIONS OF THE CUSTOMERS WHO BOUGHT THE SUV BUT PREDICTED NOT TO
3 IS THE INCORRECT PREDICTIONS OF THE CUSTOMERS WHO DIDN'T BOUGHT THE SUV BUT WERE PREDICTED TO BUY IT 
19 IS THE CORRECT NO. OF PREDICTION OF CLASS ONE(CUSTOMERS WHO BOUGHT THE SUV)
'''

accuracy_score(ytest, ypred)
```

VISUALISING THE TRAINING SET RESULTS

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(xtrain), ytrain
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

VISUALISING THE TEST SET RESULTS

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(xtest), ytest
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
